```python
def getmin(Arr,n1,n2):
    if n2-n1 == 0:
       return Arr[n2]
    else:
       k = int((n2 + n1)/2)
       return min(getmin(Arr,n1,k), getmin(Arr,k+1,n2))
```

```python
A = [-8,5,7]
getmin(A,0,2)
```

```python
def getMin(A,L,R):
    arr=A
    left= L
    right= R
    if left == right :
        return arr[left]
    if right - left == 1:
        if arr[right] > arr[left]:
            return arr[left]
        else:
            return arr[right]
            
    if right-left>1 :
        mid= int((left+right)/2)
        Lmin= getMin(arr,left,mid)
        Rmin= getMin(arr,mid+1,right)
        return min(Lmin,Rmin)
        
A=[1,-34,-2,67,8,-9,25,79,-30]
R= len(A) - 1;
L= 0
minimum= getMin(A,L,R)
print(minimum)
```

```python
def min(A,p,q):
    if(p<q):
        k = int((p+q)/2)
        M1 = min(A ,p,k)
        M2 = min(A,k+1,q)
    if(p==q):
        return A[p]
    
    if(M1<=M2):
        M=M1
    else:
        M=M2
    return M
```

```python
A=[1,-34,-2,67,8,-9,25,79,-30]
minimum= min(A,0,7)
print(minimum)
```
